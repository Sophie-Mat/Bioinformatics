### 1. Read the training data from http://139.91.190.186/tei/bioinformatics/LungTrain.txt and pre-process them in order to create a machine learning model with the sklearn library. You have to transpose the data and set the right label names if you read the file with pandas.

```python
import pandas as pd
```

#### Read the data:

```python
data  = pd.read_csv("http://139.91.190.186/tei/bioinformatics/LungTrain.txt", sep="\t")
```

```python
data
```

#### Data pre-processing for machine learning
##### Creating a list with the two sample classes, removing any numbering they have:

```python
labels = data.columns[1:]
labels
```

```python
for x in range(len(labels)):
    if( labels[x].startswith("Cancer")):
       labels.values[x] = "Cancer"
    else:
       labels.values[x] = "Normal"
labels
```

##### Remove column genres

```python
data1 = data.iloc[:, 1:]
data1
```

##### Transpose the dataset:

```python
data1 = data1.T
data1
```

### 2. Train a SVM model with the training data from step 1 and evaluate your model with the test data that you can find here http://139.91.190.186/tei/bioinformatics/LungTest.txt . For the training you will use the full dataset, no split is needed. Print the accuracy and the confusion matrix
##### Read the test file:

```python
test_data  = pd.read_csv("http://139.91.190.186/tei/bioinformatics/LungTest.txt", sep="\t")
test_data
```

##### Pre-process the test data:

```python
test_labels = test_data.columns[1:]
test_labels
```

```python
for x in range(len(test_labels)):
    if( test_labels[x].startswith("Cancer")):
       test_labels.values[x] = "Cancer"
    else:
       test_labels.values[x] = "Normal"
test_labels
```

```python
test_data1 = test_data.iloc[:, 1:]
test_data1
```

```python
test_data1 = test_data1.T
test_data1
```

##### So this is what we know after the pre-processing of the two datasets:

```python
X_train = data1
y_train = labels
X_test = test_data1
y_test = test_labels
```

##### Train a SVM model with the training data:

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
```

##### The accuracy:

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, clf.predict(X_test))
```

##### The confusion matrix:

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, clf.predict(X_test))
```
