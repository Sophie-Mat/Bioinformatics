### 1. Use the first 100 nucleotides from the DNA sequence of CTCF gene for human and mouse to create a Dot Matrix with k-mer k=6. 

#### Creating the DNA sequences of CTCF gene for both human and mouse.

```python
sequence_dna_homo = ''
for i in dna_homo_lines[1:]:
    sequence_dna_homo += i.strip()
#sequence_dna_homo
```

```python
sequence_dna_mus = ''
for i in dna_mus_lines[1:]:
    sequence_dna_mus += i.strip()
#sequence_dna_mus
```

#### Each secuence will contain 100 nucleotides

```python
sequence_dna_homo=sequence_dna_homo[0:100]
sequence_dna_mus=sequence_dna_mus[0:100]
```

#### Define the delta function:

```python
def delta(x,y):
    return 0 if x == y else 1
```

#### M[i,j]

```python
def M(seq1,seq2,i,j,k):
    return sum(delta(x,y) for x,y in zip(seq1[i:i+k],seq2[j:j+k]))
```

#### Define the creation of the Dot Matrix

```python
def makeMatrix(seq1,seq2,k):
    n = len(seq1)
    m = len(seq2)
    return [[M(seq1,seq2,i,j,k) for j in range(m-k+1)] for i in range(n)]
```

#### Plot the matrix

```python
def plotMatrix(M, seq1, seq2, t=1, nonblank = chr(0x25A0), blank = ' '):
    print(' ' + seq2)
    for label,row in zip(seq1,M):
        line = ''.join(nonblank if s < t else blank for s in row)
        print(label + line)
```

```python
def dotplot(seq1,seq2,k = 6):
    M = makeMatrix(seq1,seq2,k)
    plotMatrix(M, seq1,seq2)
```

```python
dotplot(sequence_dna_homo,sequence_dna_mus)
```

### Using the datasets from part 2, implement your own Smith-Waterman Algorithm for global alignment of the Amino Acid sequences with the following parameters: Identical = +2, substitution = -2 and gap penalty = -1. Print the final scoring matrix and find the position of the highest value. 
####  Final scoring matrix using Smith-Waterman Algorithm:

```python
import numpy as np

def matrix(a, b, match_score=2, gap_cost=1):
    H = np.zeros((len(a) + 1, len(b) + 1), np.int)

    for i in range(1, H.shape[0]):
        for j in range(1, H.shape[1]):
            match = H[i - 1, j - 1] + (match_score if a[i - 1] == b[j - 1] else - match_score)
            delete = H[i - 1, j] - gap_cost
            insert = H[i, j - 1] - gap_cost
            H[i, j] = max(match, delete, insert, 0)
    return H

print(matrix(sequence_aa_homo, sequence_aa_mus))
```

#### Final scoring matrix (a better-looking version of the above).

```python
import pandas as pd

arr_seq_aa_mus = list(sequence_aa_mus)
arr_seq_aa_mus.insert(0," ")

table = pd.DataFrame(np.array(matrix(sequence_aa_homo, sequence_aa_mus)),
                   columns=arr_seq_aa_mus)

arr_seq_aa_homo = list(sequence_aa_homo)
arr_seq_aa_homo.insert(0," ")

table.index = arr_seq_aa_homo
table
```

#### Same matrix as above but with numbers in each row and column, in order to find the position of the highest value.

```python
table_new = pd.DataFrame(np.array(matrix(sequence_aa_homo, sequence_aa_mus)))

table_new
```

#### The highest value:

```python
max_Value = max(table_new.max())
max_Value
```

#### X and Y position of the highest value:

```python
xpos=-1
ypos=-1
for i in range(101):
    for j in range(101):
        if (table_new[i][j]==max_Value):
            xpos=i
            ypos=j
            break
        
if (xpos==-1|ypos==-1):
    print("Something went wrong")
    
print("X position:", xpos,"\nY position:", ypos)
```
